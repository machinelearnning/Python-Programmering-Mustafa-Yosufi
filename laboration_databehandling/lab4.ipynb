```python
import pandas as pd
college = pd.read_csv("College.csv")
```

```python
college2 = pd.read_csv("College.csv", index_col=0)
college3 = college.rename({"Unnamed: 0": "College"}, axis = 1)
college3 = college3.set_index("College")
college = college3
```

```python
college.describe()
```

```python
C = ["Apps", "Accept", "Enroll"]
pd.plotting.scatter_matrix(college[C])
```

```python
college.boxplot(column='Outstate', by='Private')
```

```python
college["Elite"] = pd.cut(college["Top10perc"]/college["Enroll"], [0 ,0.5 ,1], labels=["No" ,"Yes"])
print(college["Elite"].value_counts())
college.boxplot(column='Outstate', by="Elite")
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 3, figsize = (15, 5))

axes[0].hist(college["Personal"], bins = 50)
axes[1].hist(college["Accept"], bins = 40)
axes[2].hist(college["Top25perc"], bins = 30)

plt.show()
```
