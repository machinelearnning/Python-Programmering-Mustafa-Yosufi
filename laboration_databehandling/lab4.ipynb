```python
import pandas as pd
college = pd.read_csv("College.csv")
college.columns
```

```python
college2 = pd.read_csv("College.csv", index_col=0)
college3 = college.rename({"Unnamed: 0": "College"}, axis = 1)
college3 = college3.set_index("College")
college = college3
```

```python
college.columns
```

```python
import matplotlib.pyplot as plt

C = ["Apps", "Accept", "Enroll"]
pd.plotting.scatter_matrix(college[C])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#college.boxplot(column= ["Private", "Outstate"])
sns.boxplot(x = "Private", y = "Outstate", data = college)
plt.show()
```

```python
college.boxplot(column='Outstate', by='Private')
```

```python
college["Elite"] = pd.cut(college["Top10perc"]/college["Enroll"], [0 ,0.5 ,1], labels=["No" ,"Yes"])
print(college["Elite"].value_counts())
college.boxplot(column='Outstate', by="Elite")
```

```python

fig, axes = plt.subplots(1, 3, figsize = (15, 5))

axes[0].hist(college["Personal"], bins = 30)
axes[1].hist(college["Accept"], bins = 30)
axes[2].hist(college["Top25perc"], bins = 30)

plt.show()
```

```python
plt.figure(figsize= (15, 5))
plt.subplot(1, 2, 1)
plt.hist(college["Personal"], bins = 30)
plt.title("Personal")

plt.subplot(1, 2, 2)
plt.hist(college["Accept"], bins = 30)
plt.title("Accept")
plt.show()
```
