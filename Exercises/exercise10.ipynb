```python
# 1. Dice simulations
import numpy as np
import matplotlib.pyplot as plt
sex_side = np.array([range(1, 7)])
sex_side.mean()

dic_num = [10, 100, 1000, 10_000, 100_000, 1000_000]
dic_mean = [ np.random.randint(1, 7, size=x).mean() for x in dic_num]
plt.plot(dic_mean, "*-")
plt.xticks([0,1,2,3,4,5], dic_num) # istället för dic_means indexer lägg dic_num
```

```python
# 2. Several dices

x = np.random.randint(1, 7, 6)
y = np.random.randint(1, 7, 6)
lis = []
for x in range(6):
    lis.append(np.add(x, y))

unique_values, dd = np.unique(lis, return_counts= True)
probibility = [round(x/36, 3) for x in dd]

plt.bar(unique_values, probibility)
```
