```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv("unlabelled_data.csv")

df.columns = df.columns.str.replace("-1.885907518189583", "x").str.replace("-1.997407599218205", "y")
x_points = [-4, 4]
y_points = [-4, 4]

df.plot(kind = "scatter", x = "x", y = "y")
plt.plot(x_points, y_points, "r")
#ee = [x_points[x] - int(x_points[x]) for x in x_points]
def k_m_values(x_points, y_points): #find m and k values
    lis = [1 for i in range(x_points[0], x_points[1])]
    lis1 = [1 for i in range(y_points[0], y_points[1])]
    global k_value
    k_value = sum(lis1)/sum(lis)
    global m_value
    m_value = y_points[0] + (-1 * x_points[0] * k_value)
    return k_value, m_value

k_m_values(x_points, y_points)
```

```python

y_above = []
y_under = []
x_above = []
x_under = []
for x, y in zip(df["x"], df["y"]):
    y_value1 = k_value*x + m_value
    if y > y_value1:
      y_above.append(y)
      x_above.append(x)
    else:
        y_under.append(y)
        x_under.append(x)

plt.scatter(x = x_above, y = y_above, color = "red")
plt.scatter(x = x_under, y = y_under, color = "orange")
plt.show()
```

```python

df["label"] = df["x"]

for x, y in zip(df["x"], df["y"]):
    y_value = k_value*x + m_value
    if y > y_value:
      df["label"] = df["label"].replace(x, 0)
    else:
       df["label"] = df["label"].replace(x , 1)
    
df["label"] = df["label"].astype(int)
```

```python
labels_0 = df[df["label"] == 0]
labels_1 = df[df["label"] == 1]

plt.scatter(x = labels_0["x"], y = labels_0["y"], color = "red")
plt.scatter(x = labels_1["x"], y = labels_1["y"], color = "orange")
plt.plot(x_points, y_points, "b")
plt.show()
```
